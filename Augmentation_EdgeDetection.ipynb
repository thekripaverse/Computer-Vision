```python
img_s=img[10:110,50:225]
cv2_imshow(img_s)
```

```python
img_c90=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
img_a90=cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(img_c90)
cv2_imshow(img_a90)
```

```python
img_flip=cv2.flip(img,0)
cv2_imshow(img_flip)
```

```python
img_hflip=cv2.flip(img,1)
cv2_imshow(img_hflip)
```

```python
cv2.resize(img,(100,100))
```

```python
img_blur=cv2.blur(img,(10,10))
cv2_imshow(img_blur)
```

```python
img_padding=cv2.copyMakeBorder(img, 20,20,20,20, cv2.BORDER_CONSTANT, value=[0,0,0])
cv2_imshow(img_padding)
```

```python
img_bright=cv2.convertScaleAbs(img,alpha=2,beta=100)
cv2_imshow(img_bright)
```

```python
noise = np.random.normal(200,100, img.shape)
noisy_image = img.astype(float) + noise
cv2_imshow(noisy_image)
```

```python
M= np.float32([[1, 0.5, 0],[0, 1,   0]])
sheared_image = cv2.warpAffine(img,M,(225,225))
cv2_imshow(sheared_image)
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv_img)
```

```python
M = np.float32([[1, 0, 50], [0, 1, 100]])
translated_img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
cv2_imshow(translated_img)
```

EDGE DETECTION

```python
gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edge=cv2.Canny(gray_image,30,200)
cv2_imshow(img)
cv2_imshow(edge)
```
